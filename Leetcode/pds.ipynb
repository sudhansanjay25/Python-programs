NUMPY SIPLE IMPLEMENTATION

```python
import numpy as np
n = int(input())
l = [int(x) for x in input().split()[:n]]
l = np.array(l)
print(l.__lt__(6))
print(l.__gt__(6))
print(l.__eq__(6))
print(l.__le__(6))
print(l.__ge__(6))
print(l.__ne__(6))
print(np.negative(l))
```

PANDAS INDEXING AND SELECTION

```python
import pandas as pd
data = pd.Series([0.25,0.50,0.75,1.0],['a','b','c','d'])
print(data)
print(data[1])
print(data.iloc[1])
print(data.iloc[1:3])
```

```python
import numpy as np
import matplotlib.pylab as plt
from scipy.stats import norm
import statistics
x = np.arange(-20, 20, 0.01)
mean = statistics.mean(x)
sd = statistics.stdev(x)
plt.plot(x, norm.pdf(x, mean, sd))
plt.title("Curve")
plt.show()
```
