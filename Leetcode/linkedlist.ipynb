```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def ad(self, data):
        temp = Node(data)
        if not self.head:
            self.head = temp  # Set the head for the first node
            return
        last = self.head
        while last.next:
            last = last.next
        last.next = temp

    def printt(self):
        temp = self.head
        while temp:
            print(temp.data, end=" ")
            temp = temp.next
        print()  # New line after the list is printed

    def check(self):
        d = self.head
        s = []  # Using list instead of set since order matters
        while d:
            if d.data not in s:  # Prevent adding duplicates
                s.append(d.data)
            d = d.next
        temp = LinkedList()
        temp.ad(s[0])
        for i in range(1, len(s)):
            temp.ad(s[i])
        return temp

# Create the linked list and add elements
l = LinkedList()
l.ad(1)
l.ad(2)
l.ad(2)
l.ad(3)
l.ad(3)
l.ad(4)
l.ad(5)

print("Original List:")
l.printt()

# Remove duplicates and print the result
print("List after removing duplicates:")
l_no_duplicates = l.check()
l_no_duplicates.printt()
```

```python
class node:
    def __init__(self, coef, p):
        self.coef = coef
        self.p = p
        self.next = None
        self.prev = None
        
class linkedlist:
    def __init__(self):
        self.h1 = None
        self.t1 = None
        self.h2 = None
        self.t2 = None
        self.t3 = None
        self.h3 = None
        
    def add(self, coef, p , head, tail):
        new = node(coef, p)
        if not head:
            head = tail = new
        else:
            tail.next = new
            new.prev = tail
            tail = new
        return head, tail
        
    def display(self, head):
        while head!=None:
            if head.next != None:
                c = None
                if head.next.coef > 0:
                    c=" + "
                else:
                    c =" - "
            coef = abs(head.coef)
            if head.next != None:
                if head.p != 1 and head.p != 0:
                    print(coef,"x",head.p,sep = "", end=c)
                elif head.p == 1:
                    print(coef,"x", sep ="",end=c)
                else:
                    print(head.coef,end=c)
            else:
                if head.p != 1 and head.p != 0:
                    print(coef,"x",head.p,sep = "")
                elif head.p == 1:
                    print(coef,"x", sep ="")
                else:
                    print(coef)
            head = head.next
            
l = linkedlist()
a = int(input())
l1 = [int(c) for c in input().split()[:a]]

b = int(input())
l2 = [int(x) for x in input().split()[:b]]

for i in range(0,len(l1),2):
    l.h1, l.t1 = l.add(l1[i],l1[i+1],l.h1, l.t1)
    
for i in range(0, len(l2), 2):
    l.h2, l.t2 = l.add(l2[i], l2[i+1], l.h2, l.t2)
    
curr, temp = l.h1, l.h2

while curr != None and temp != None:
    a = None
    po = None
    if curr.p == temp.p:
        a = curr.coef + temp.coef
        po = curr.p
        curr = curr.next
        temp = temp.next
    elif curr.p > temp.p:
        a = curr.coef
        po = curr.p
        curr = curr.next
    else:
        a = temp.coef
        po = temp.p
        temp = temp.next
    l.h3, l.t3 = l.add(a, po, l.h3, l.t3)
while curr != None:
    l.h3, l.t3 = l.add(curr.coef, curr.p, l.h3, l.t3)
while temp != None:
    l.h3, l.t3 = l.add(temp.coef, temp.p, l.h3, l.t3)
l.display(l.h3)
```
