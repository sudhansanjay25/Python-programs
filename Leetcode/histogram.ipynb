## **Histogram**

```python
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
x = [10,20,20,30,30,30,40,40,40,40,50,50,50,50,50]
plt.hist(x,bins=80,color="green",orientation='vertical')
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.title("Student")

plt.show()
```

```python
x = [10,20,20,30,30,30,40,40,40,40,50,50,50,50,50]
plt.hist(x,bins=80,color="orange",orientation='horizontal')
plt.xlabel("Frequency")
plt.ylabel("Marks")
plt.title("Student")

plt.show()
```

```python
import librosa
import librosa.display
import matplotlib.pyplot as plt

# Load an audio file
audio_file = r"C:/Users/Lenovo/Downloads/audio track.mp3"
y, sr = librosa.load(audio_file)

# Create a log-Mel spectrogram
S = librosa.feature.melspectrogram(y, sr=sr, n_mels=128)
log_S = librosa.power_to_db(S, ref=np.max)

# Display the spectrogram
plt.figure(figsize=(10, 4))
librosa.display.specshow(log_S, sr=sr, x_axis='time', y_axis='mel')
plt.colorbar(format='%+2.0f dB')
plt.title('Log-Mel Spectrogram')
plt.tight_layout()
plt.show()
```
